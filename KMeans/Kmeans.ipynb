```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# 设置风格
plt.style.use('seaborn')
```

```python
# 生成数据
n_feature = 2
centers = [[-1, -1], [0, 0], [1, 1], [2, 2]]
cluster_std = [0.4, 0.2, 0.3, 0.2]
X,Y = make_blobs(n_samples=1000, n_features=n_feature, centers=centers, cluster_std=cluster_std, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
```

```python
# 学习曲线
l = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
    l.append(kmeans.inertia_)
plt.plot(range(2, 11), l)
plt.plot(range(2, 11), l, 'o')
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
```

```python
y_pred = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y_pred, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black')
```

### 轮廓系数

```python
silhouette_score(X, y_pred)
```

轮廓系数>0.6,模型效果很好
